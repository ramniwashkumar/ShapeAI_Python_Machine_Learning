```python
import numpy as np
import pandas as pd
import sklearn

from sklearn.datasets import load_boston
df = load_boston()

df.keys()
#output dict_keys(['data','target','feature_names','DESCR'])
print(df.DESCR)
print(df.feature_name)
print(df.target)

boston = pd.DataFrame(df.data, columns=df.feature_names)
boston.head(20)
#print first 20 line of dataset by default first 5 line of data

#adding a new column of target values to dataframe
boston['TARG'] = df.target
boston.head()

#Check if dataset contains any null value or not
boston.isnull()

#if any null value is present in particular column
boston.isnull().sum()
#return no. of null values in that column

from sklearn.model_selection import train_test_split

X = boston.drop('TARG',axis=1)
#x-axis(axis=0), y-axiss(axis=1)

Y = boston['TARG']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.15, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)


from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lin_model=LinearRegression()
lin_model.fit(X_train, Y_train)

y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))

print("The model performance for training set")
print('RMSE is {}'.format(rmse))
print("\n")

#on testing set
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
print("The model performance for testing set")
print('RMSE is {}'.format(rmse))


```
